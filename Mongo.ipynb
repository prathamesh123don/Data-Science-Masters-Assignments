##### Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use MongoDB over SQL databases?

-> MongoDb is a document oriented NoSQL database. These are the database that do not follow the traditional relational database model. Instead they store data in different ways, such as in documents key value pairs or graphs.

It stores the data in documents , which are JSON like objects. This makes it easy to store query data that is not well suited for a relational database, such as semi-structured or unstructured data.

Benifits are
1. Flexibility.
2. Scalability.
3. Performances.

The scenarios where it is prefered to use MongoDB over SQL databases:

1. Storing semi-structured or unstructured data.
2. Highly scalabe.
3. Develop applications quickly. 

It does not supports advance features such as joins and subqueries.

##### Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.

-> Sure, the sort() method is used to sort the results of a query in MongoDB. It takes a document as an argument, which specifies the fields that the documents should be sorted by and the order in which they should be sorted. 

import pymongo

<!-- # Create a MongoClient object -->
client = pymongo.MongoClient()

<!-- # Connect to the "mydb" database -->
db = client["mydb"]

<!-- # Create a collection named "mycollection" -->
collection = db["mycollection"]

<!-- # Insert a document into the collection -->
document = {"name": "Bard", "age": 10000}
collection.insert_one(document)

<!-- # Insert another document into the collection -->
document = {"name": "Alice", "age": 20}
collection.insert_one(document)

<!-- # Sort the documents in the collection by the "age" field in ascending order -->
cursor = collection.find().sort({"age": 1})

<!-- # Iterate over the cursor and print the documents -->
for document in cursor:
    print(document)


##### Q7. Explain why delete_one(), delete_many(), and drop() is used.

->  The delete_one(), delete_many(), and drop() methods are used to delete documents from a MongoDB collection.

The delete_one() method deletes a single document from a collection.
The delete_many() method deletes multiple documents from a collection.
The drop() method deletes all documents from a collection.
